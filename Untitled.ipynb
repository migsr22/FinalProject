```python
# Function to get the column index from column name
def column_index(df, query_cols):
    cols = df.columns.values
    sidx = np.argsort(cols)
    return sidx[np.searchsorted(cols,query_cols,sorter=sidx)]
```

```python
# Where are our outcome variables?
column_index(df, ['death', 'gstatus_ki'])
```

```python
# Appending columns to the end of our dataframe
cols = list(df.columns.values) #Make a list of all of the columns in the df

cols.pop(cols.index('death')) #Remove death from list
cols.pop(cols.index('gstatus_ki')) #Remove gstatus_ki from list
df = df[cols+['death','gstatus_ki']] #Create new dataframe with columns in the order
```

```python
# Where are out outcome variables now ?
column_index(df, ['death', 'gstatus_ki'])
```

```python
# Identify the amount of missing values we have for each column
datadict = df.isnull().sum().plot(kind='bar')
plt.title('Total Missing Values')
plt.xlabel('Columns')
plt.ylabel('Missing Values')
plt.show()

# Identify the amount of missing values we have for each column, option #2
# datadict['Missing Values'] = df.isnull().sum()
# datadict
```

```python
# Drop columns that have more than 20% NaN values
df = df.loc[:, df.isnull().mean() < .20]
```

```python
# Visualizing the missing values after dropping NaN columns (20%)
datadict = df.isnull().sum().plot(kind='bar')
plt.xlabel('Columns')
plt.ylabel('Missing Values')
plt.ylim((0,250000))
plt.show()
```

```python
# Identify datatypes of the Columns, add the stats to the datadict
datadict = pd.DataFrame(df.dtypes)
datadict['Missing_Values'] = df.isnull().sum()
datadict['Num_Unique'] = df.nunique()
datadict['Count']= df.count()
datadict = datadict.rename(columns={0:'DataType'})
datadict
```

```python
# Let's take a loot at death rate in our df
# df['death'].value_counts()
# 0    177834 survived                      
# 1     62545 die                 
# Survival rate (177834/240379) = ~74% 
fig, axs = plt.subplots(1,2)
df['death'].value_counts().plot(kind='bar', ax=axs[0], title = "Total count of Survivorship")
plt.ylabel("Total")
plt.xlabel("Survived          Died")
df['death'].value_counts(normalize=True).plot(kind='bar', ax=axs[1], title = "Rates of Survivorship")
plt.ylabel("Rate")
plt.xlabel("Survived          Died")
plt.show()
```
